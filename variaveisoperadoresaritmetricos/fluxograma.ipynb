# Hora extra
image.png

```python
hora = int(input('Digite as horas trabalhadas: '))
```

```python
min = int(input("Digite os minutos trabalhados: "))
```

```python
valorhora = float(input('Digite quanto você ganha por hora: '))
```

```python
conver = hora+(min/60)
salario = (conver*valorhora)
resultado = (salario*70/100)+salario
```

```python
print(f"O valor pelas horas trabalhadas será de: {resultado:.2f}")
```
